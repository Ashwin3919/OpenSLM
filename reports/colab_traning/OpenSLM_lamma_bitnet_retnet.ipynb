```python
import subprocess, multiprocessing, os

models = ["train-mamba", "train-rwkv", "train-jamba"]

def train(target):
    model_name = target.replace("train-", "")
    log_file = f"logs_3/{model_name}.log"
    os.makedirs("logs_3", exist_ok=True)

    env = os.environ.copy()
    env["PYTORCH_CUDA_ALLOC_CONF"] = "max_split_size_mb:512"

    with open(log_file, "w") as f:
        subprocess.run(
            f"cd OpenSLM-main && make {target}",
            shell=True, env=env,
            stdout=f, stderr=f
        )
    print(f"[DONE] {model_name} → {log_file}")

if __name__ == "__main__":
    os.makedirs("logs_3", exist_ok=True)
    with multiprocessing.Pool(3) as pool:  # ← fixed: 3 not 6
        pool.map(train, models)
```
