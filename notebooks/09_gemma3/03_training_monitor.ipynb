# 03 — Gemma 3 Training Monitor

Visualise loss curves and learning rate schedules from Gemma 3 experiments.

```python
import sys
sys.path.insert(0, '../..')

import json
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
```

```python
METRICS_PATH = Path('../../outputs/gemma3/metrics.json')

if METRICS_PATH.exists():
    with open(METRICS_PATH) as f:
        metrics = json.load(f)
    
    df = pd.DataFrame(metrics)
    
    plt.figure(figsize=(10, 4))
    plt.plot(df['iteration'], df['train_loss'], label='Train Loss')
    plt.plot(df['iteration'], df['val_loss'], label='Val Loss')
    plt.title('Gemma 3 — Loss Curves')
    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print(f'No metrics found at {METRICS_PATH}. Start training first!')
```
