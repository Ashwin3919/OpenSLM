# 03 — Training Monitor

Reads `outputs/metrics.json` written by `TrainingPipeline.save_results()`
and plots train vs. validation loss curves.  Works mid-training — just
re-run the cells to refresh.

```python
import json
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ── Load metrics ──────────────────────────────────────────────────────────
metrics_path = Path('../../outputs/retnet/metrics.json')

if not metrics_path.exists():
    print(f'Not found: {metrics_path}  (start training first with make train)')
else:
    with open(metrics_path) as f:
        metrics = json.load(f)

    train_losses = metrics['train_losses']
    val_losses   = metrics['val_losses']
    interval     = metrics['eval_interval']
    print(f'Evaluation points: {len(train_losses)}')
    print(f'Best val loss: {metrics["best_val_loss"]:.4f}')
```

```python
# ── Loss curves ───────────────────────────────────────────────────────────
if metrics_path.exists() and train_losses:
    steps = [i * interval for i in range(1, len(train_losses) + 1)]

    plt.figure(figsize=(10, 5))
    plt.plot(steps, train_losses, 'g-o', markersize=4, label='train loss')
    plt.plot(steps, val_losses,   'r-o', markersize=4, label='val loss')
    plt.xlabel('Iteration')
    plt.ylabel('Cross-entropy loss')
    plt.title('Training progress')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```
